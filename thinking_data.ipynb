```python
import json
```

```python
#load data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
#function for clean and structure data

def clean_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        #clean rating
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating) in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # handle missing vlaue
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        # Deduplicaation
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
        
```

```python
data = clean_data(data)
```

```python
#Get meaningful insights from data

def get_insights(data):

    #avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])
    print(f"average ={tot_rating/len(data)}")

    # % of user of poor rating
    poor_ratings = 0
    for user in data:
        if(float(user["rating"]) <3):
            poor_ratings += 1
    print(f"% of poor rating of user rating = {poor_ratings/len(data) * 100}")
    
```

```python
get_insights(data)
```

```python
#Recomendation features   

def get_recomendations(data):
    recomendations = []

    for user in data:
        curr_recomm= {}
        curr_recomm["name"] = user["name"]
        if(float(user["rating"]) >=4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung "
        recomendations.append(curr_recomm)
    
    return recomendations 
```

```python
get_recomendations(data)
```
