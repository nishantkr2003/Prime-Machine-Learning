# FIbonacci Serise

```python
n = 10
a = 0
b = 1
for i in range(n):
    print(a, end = ",")
    temp = a
    a = b
    b = temp + b
```

```python
n = 10
a = 0
b = 1

for i in range(n):
    print(a, end=", ")
    a, b = b, a + b
```

```python
def fibonacci(n):
    if n<=1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

for i in range(10):
    print(fibonacci(i) , end=" ")
        
```

```python
n = 10
a = 0
b = 1

count = 0
while count<n:
    print(a, end=" ")
    a,b = b,a+b
    count +=1
    
    
```

# Palindrome

```python
a = "12321"

if a == a[::-1]:
    print("p")
else:
    print("Not P")
```

```python
a = "122231"

rev = ""
for i in a:
    rev = i + rev

if(a == rev):
    print("P")
else:
    print("Not p")
```

```python
a = "12213"
rev = a[::-1]

if(a == rev):
    print("P")
else:
    print("Not p")
```

# Count Letter digits and spaces 

```python
text = "Hello 123 World"

letters = 0
digits = 0
spaces = 0

for char in text:
    if char.isalpha():
        letters += 1
    elif char.isdigit():
        digits += 1
    elif char.isspace():
        spaces += 1

print(f"Letters : {letters}\nDigits : {digits} \nSpaces  : {spaces}")
```

# counting special cheracter in a string

```python
def count_sp_char(string):
    sp_char = "!@#$%^&*(){}[]|/<>"
    count = 0
    for i in string:
        if i in sp_char:
            count += 1
    return count
text = "Hello@123 #World!"
print(count_sp_char(text))
```
